```python
def obtener_generos():
    url = "https://books.toscrape.com/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encuentra la sección de géneros; usualmente está en un elemento <ul> con clase 'nav' o similar
    category_section = soup.find('ul', class_='nav-list')
    
    genres = []
    if category_section:
        genre_links = category_section.find_all('a')
        for link in genre_links:
            genre_name = link.text.strip()
            # Evitar incluir el Género 'Books' general
            if genre_name.lower() != 'books':
                genres.append(genre_name)
    
    return genres

# Llamada a la función
listado_generos = obtener_generos()
#print("Géneros Disponibles:")
#for genre in listado_generos:
#    print(genre)

print(listado_generos)
```

```python
def scrape_books(min_rating, max_price):
    # Inicializar una lista para almacenar los datos de libros
    books_data = []
    
    # Define la URL
    url = "https://books.toscrape.com/catalogue/page-{}.html"

    pag = int(input(f'Por favor, introduce cuántas páginas tiene el catálogo'))
    for page in range(1, pag):  # Ajusta el rango según las páginas disponibles
        response = requests.get(url.format(page))
    
    soup = BeautifulSoup(response.text, 'html.parser')
       

    for article in soup.find_all('article', class_='product_pod'):
        # Continuar con la extracción de datos por libro
        
        book_url = article.h3.a['href']
        book_page = requests.get("https://books.toscrape.com/catalogue/" + book_url)
        book_soup = BeautifulSoup(book_page.content, 'html.parser')
        # Asumamos que el género está en una lista de breadcrumbs en la página de detalles del libro
        breadcrumb = book_soup.find('ul', class_='breadcrumb')
        if breadcrumb:
            genre_link = breadcrumb.find_all('li')[2].a  # Ajusta el índice para saltar "Home" y "Books"
            genre = genre_link.text.strip()
        else:
            genre = 'Unknown'  # Fallback si no se encuentra

        title = article.h3.a['title']
        price = article.find('p', class_='price_color').text
        price = float(price[2:])  # Convertir a float, omitir símbolo "£"

        rating_class = article.p['class'][1]
        rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
        rating = rating_map.get(rating_class, 0)

        availability = article.find('p', class_='instock availability').text.strip()

        book_url = article.h3.a['href']
        book_page = requests.get("https://books.toscrape.com/catalogue/" + book_url)
        book_soup = BeautifulSoup(book_page.content, 'html.parser')

        upc = book_soup.find('th', string='UPC').find_next('td').text

        description_tag = book_soup.find('meta', {'name': 'description'})
        description = description_tag['content'].strip() if description_tag else 'No description available'

        if rating >= min_rating and price <= max_price:
            books_data.append({
            'UPC': upc,
            'Title': title,
            'Price (£)': price,
            'Rating': rating,
            'Genre': genre, 
            'Availability': availability,
            'Description': description,
            })

    return pd.DataFrame(books_data)

df = scrape_books(min_rating=4, max_price=20)
df
```
